Project Summary: Exploratory Data Analysis of Ford GoBike System
1. Introduction & Context
The micro-mobility landscape has undergone a massive transformation, shifting from a niche alternative to a core pillar of urban transit systems. This project centers on conducting a thorough Exploratory Data Analysis (EDA) of the Ford GoBike sharing system using Google Colab. The primary objective was to analyze raw, multi-file transactional trip data to uncover hidden patterns in user behavior, optimize operational logistics, evaluate social equity programs, and deliver actionable data-driven strategies to maximize the platform's overall business impact.

2. Data Pipeline & Wrangling Architecture
The project began by building a robust data integration pipeline. Because the raw data was split across multiple individual comma-separated files (CSVs), Python's glob and pandas libraries were used to dynamically locate, parse, and merge the datasets into a single master DataFrame. Once unified, the data underwent a comprehensive data cleaning process to ensure analytical integrity. High-contrast missing value heatmaps revealed null values within critical geographic parameters, such as station names and structural coordinates, which were handled systematically to prevent data bias.
Type casting was performed to convert string timestamps (start_time and end_time) into standard Python datetime objects. From these cleaned baselines, extensive feature engineering was performed to extract granular temporal insights, generating new analytical fields including start_hour, day_of_week, and month. Finally, trip durations were standardized from raw seconds into duration_min to provide a more intuitive, human-readable format for statistical modeling.

3. Core Insights & Analytical Discoveries
The exploratory phase evaluated 15 distinct Univariate, Bivariate, and Multivariate visualizations, revealing a clear contrast between user segments:
Demographic Dominance: Annual Subscribers form the backbone of the platform, driving approximately 85% to 90% of total trip volume. This segment consists largely of working professionals and students aged 25–38.
Temporal Variations: System demand follows an intense bimodal distribution on weekdays, peaking sharply during morning commute hours (8:00 AM – 9:00 AM) and evening rush hours (5:00 PM – 6:00 PM). Subscribers drive these spikes, using the fleet for predictable, time-sensitive transit.
Leisure Dynamics: Conversely, casual Customers display a smooth, bell-shaped midday usage curve. While weekend volumes drop overall, the average trip duration shifts dramatically—increasing by roughly 60% compared to weekdays. Casual riders keep bikes out much longer for open-ended leisure and tourist travel.
Structural Anomalies: Multi-variable pair plots and correlation heatmaps confirmed that while temporal indices operate independently, ride durations remain highly dependent on user classification.

4. Strategic Business Solutions & Conclusion
The insights gained from this EDA translate directly into four key strategic recommendations for the business:
Smart Fleet Rebalancing: Ground crews can minimize dock shortages by running bike redistribution runs during off-peak windows (10:00 AM to 2:00 PM), moving inventory toward commercial zones ahead of the evening rush.
Targeted Demographics Expansion: Marketing efforts should focus on closing the platform's gender gap (where male riders currently make up over 70% of trips) by promoting safety, neutral lifestyle benefits, and protected municipal bike infrastructure.
Weekend Monetization: Introducing "Weekend Explorer Passes" or local commercial partnerships can successfully monetize idle weekend inventory.
Simplified Equity Programs: Streamlining onboarding for the "Bike Share for All" program will drive community adoption and secure valuable municipal subsidies.

```python
import os
import zipfile
import glob
import pandas as pd

# Define paths matching the folder you created in Drive
drive_folder_path = '/content/drive/MyDrive/Ford-Go-bike sharing'
extract_to_path = '/content/extracted_bike_data'

# Create a local directory in Colab environment to extract files into
os.makedirs(extract_to_path, exist_ok=True)

# 1. Look for all zip files in your Drive folder
zip_files = glob.glob(os.path.join(drive_folder_path, '*.zip'))

print(f"Found {len(zip_files)} zip files. Extracting...")
for zip_file in zip_files:
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_to_path)
print("Extraction complete!")

# 2. Find all the extracted CSV files and combine them
csv_files = glob.glob(os.path.join(extract_to_path, '*.csv'))
df_list = []

print(f"Found {len(csv_files)} CSV files. Merging into one dataframe...")
for file in csv_files:
    temp_df = pd.read_csv(file)
    df_list.append(temp_df)

# The final combined dataset
# Check if df_list is empty before concatenation to prevent ValueError
if not df_list:
    print("Warning: No CSV files were found and merged. Creating an empty DataFrame.")
    df = pd.DataFrame() # Create an empty DataFrame if no files were found
else:
    df = pd.concat(df_list, ignore_index=True)

print(f"✅ Success! Merged dataset shape: {df.shape}")
```

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# 1. Create a deep copy of the dataframe to protect original data
df_wrangled = df.copy()

# 2. Handle missing/null values (Dropping rows missing critical station details or gender if applicable)
critical_cols = ['start_station_name', 'end_station_name']
df_wrangled = df_wrangled.dropna(subset=[col for col in critical_cols if col in df_wrangled.columns])

# 3. Convert start and end times from object strings to datetime format
if 'start_time' in df_wrangled.columns:
    df_wrangled['start_time'] = pd.to_datetime(df_wrangled['start_time'])
if 'end_time' in df_wrangled.columns:
    df_wrangled['end_time'] = pd.to_datetime(df_wrangled['end_time'])

# 4. Feature Engineering: Extract time-based metrics for visual profiling
if 'start_time' in df_wrangled.columns:
    df_wrangled['start_hour'] = df_wrangled['start_time'].dt.hour
    df_wrangled['day_of_week'] = df_wrangled['start_time'].dt.day_name()
    df_wrangled['month'] = df_wrangled['start_time'].dt.month_name()

# 5. Feature Engineering: Convert trip duration from seconds to minutes for clarity
if 'duration_sec' in df_wrangled.columns:
    df_wrangled['duration_min'] = df_wrangled['duration_sec'] / 60

# 6. Verify changes
print("Data Wrangling Complete.")
print(f"New shape after cleaning: {df_wrangled.shape}")
print(df_wrangled.info())
```

### What all manipulations have you done and insights you found?

Data Manipulations Done!

Data Preservation: Formed a clean copy (df_wrangled) to safely run operations without mutating raw data.

Null Value Handling: Isolated and removed incomplete rows missing critical location information to ensure visualization integrity.

Type Casting: Converted the string timestamps (start_time, end_time) into standard datetime objects.

Feature Extraction: Engineered new granular features including start_hour, day_of_week, and month to allow deep-dive seasonal and time-of-day analyses.

Unit Standardization: Computed duration_min from raw seconds data to provide more intuitive human-readable statistics for trip lengths.
